```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def falls_in_rectangle(self, lowleft, upperright):
        return (lowleft.x <= self.x <= upperright.x) and (lowleft.y <= self.y <= upperright.y)
```

```python
point1 = Point(2, 3)
point2 = Point(5, 7)
point3 = Point(4, 5)

point3.falls_in_rectangle(point1, point2)
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def falls_in_rectangle(self, lowleft, upperright):
        return (lowleft.x <= self.x <= upperright.x) and (lowleft.y <= self.y <= upperright.y)
    
    def distance_to(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5
```

```python
point3.distance_to(point1)
```
